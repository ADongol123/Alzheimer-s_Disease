```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.stats import ttest_ind

```

```python
dataset = pd.read_csv("./dataset.csv")
```

```python
print(dataset.info())
```

```python
dataset.head()
```

```python
dataset = dataset[dataset['Data_Value'].notnull()]
```

```python
dataset_copy = dataset.copy()  # Ensure you're working with a fresh copy

# Extract geolocation
dataset[['Longitude', 'Latitude']] = dataset['Geolocation'].str.extract(r'POINT \((-?\d+\.\d+)\s(-?\d+\.\d+)\)')

# Convert to numeric
dataset_copy.loc[:, 'Longitude'] = pd.to_numeric(dataset['Longitude'])
dataset_copy.loc[:, 'Latitude'] = pd.to_numeric(dataset['Latitude'])
```

```python
dataset_copy.head()
```

```python
# Grouping: Average data value by topic
grouped_topic = dataset_copy.groupby('Topic')['Data_Value'].mean().sort_values()
print("Average Data_Value by Topic:")
print(grouped_topic)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(dataset_copy['Data_Value'], kde=True, bins=20, color='blue')
plt.title('Distribution of Data_Value')
plt.xlabel('Data Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Heatmap of Data_Value by State and Year
pivot_table = dataset_copy.pivot_table(index='LocationDesc', columns='YearStart', values='Data_Value', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='coolwarm', annot=False)
plt.title('Data_Value Heatmap by State and Year')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
dataset_copy['Cluster'] = kmeans.fit_predict(dataset_copy[['Data_Value']].dropna())
sns.scatterplot(data=dataset_copy, x='Longitude', y='Latitude', hue='Cluster', palette='viridis')
plt.title('Clusters of States Based on Data_Value')
plt.show()
```

```python
# Group data by year and calculate average data value
trend = data.groupby('YearStart')['Data_Value'].mean().reset_index()

# Plot trends
sns.lineplot(data=trend, x='YearStart', y='Data_Value', marker='o')
plt.title("Average Data Value Over Time")
plt.xlabel("Year")
plt.ylabel("Average Data Value")
plt.show()
```

```python
# Aggregate data by location
state_avg = data.groupby('LocationDesc')['Data_Value'].mean().reset_index()

# Plot top 10 states
top_states = state_avg.sort_values('Data_Value', ascending=False).head(10)
sns.barplot(data=top_states, x='Data_Value', y='LocationDesc', palette='viridis')
plt.title("Top 10 States by Average Data Value")
plt.xlabel("Average Data Value")
plt.ylabel("State")
plt.show()
```

```python
# Aggregate by stratification categories
strat_avg = data.groupby('Stratification1')['Data_Value'].mean().reset_index()

# Plot stratification comparison
sns.barplot(data=strat_avg, x='Data_Value', y='Stratification1', palette='coolwarm')
plt.title("Average Data Value by Stratification")
plt.xlabel("Average Data Value")
plt.ylabel("Stratification")
plt.show()
```

```python
# Calculate confidence interval range
data['Confidence_Range'] = data['High_Confidence_Limit'] - data['Low_Confidence_Limit']

# Plot the confidence range
sns.histplot(data['Confidence_Range'], kde=True, bins=30, color='purple')
plt.title("Distribution of Confidence Interval Range")
plt.xlabel("Confidence Range")
plt.ylabel("Frequency")
plt.show()

```

```python
# Aggregate data by class and topic
class_topic_avg = data.groupby(['Class', 'Topic'])['Data_Value'].mean().reset_index()

# Plot data
plt.figure(figsize=(12, 8))
sns.barplot(data=class_topic_avg, x='Data_Value', y='Topic', hue='Class')
plt.title("Average Data Value by Class and Topic")
plt.xlabel("Average Data Value")
plt.ylabel("Topic")
plt.legend(title="Class")
plt.show()
```

```python
# Boxplot to identify outliers
sns.boxplot(data=data, x='Data_Value')
plt.title("Outlier Detection in Data Value")
plt.xlabel("Data Value")
plt.show()
```

```python
# Correlation heatmap
corr = data[['Data_Value', 'Low_Confidence_Limit', 'High_Confidence_Limit']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
